```python
import matplotlib.pyplot as plt

year = [2021,2022,2023,2024,2025,2026,2027]
population = [100,250,300,470,525,170,210]

plt.plot(year,population,color="red",marker="*")
plt.title("Growth of Population(Year wise)")
plt.xlabel("Years")
plt.ylabel("Population")
plt.xticks(year)  # Set x-ticks to match the years
plt.savefig("population.png",dpi=1000)
plt.show()
```
